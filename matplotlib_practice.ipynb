```python
###PART 2: BAR PLOTS ###
```

```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10,25,5,30,21]
germinated = [74, 88, 56,69,59]
```

```python
plt.figure(figsize = (6,6))
plt.bar(plants, germinated, color = 'orchid')
plt.title('Germination Rates')
plt.ylabel('Number of Seedlings')
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.bar(plants, died, color = 'red')
plt.title('Failure Rate')
plt.ylabel('Number of Dead Seedlings')
plt.show()
```

```python
# align bars by left edge

plt.figure()
plt.bar(plants, died, color = 'orchid', align = 'edge') # align bars by left edge
plt.bar(plants, germinated, color = 'olive')
```

```python
# Stack bars

plt.figure(figsize = (6,6))
plt.bar(plants, died, bottom = germinated, color = 'red')   #Define the 'bottom' parameter
plt.bar(plants, germinated, color = 'olive')
plt.show()
```

```python
# Now with all the right labels

plt.figure(figsize = (6,6))
plt.bar(plants, germinated, color = 'green', label = 'Germinated Seedlings')
plt.bar(plants, died, bottom = germinated, color = 'red', label = 'Failed Seedlings')
plt.legend(shadow=True, frameon=True, facecolor="white")
plt.title('Greenhouse Seedlings')
plt.xlabel('Plants')
plt.show()
```

```python
# Can also do horizontal bar plots

plt.figure(figsize = (8,6))
plt.barh(plants, germinated, label = 'Germinated Seedlings')                # barh command for horizontal bars
plt.barh(plants, died, left = germinated, label = 'Failed Seedlings')   # to stack, use "left" (instead of "bottom")
plt.title('Greenhouse Seedlings')
plt.legend()
plt.show()
```

```python
###Part 6: Subplots
```

```python
nums = np.arange(5)

plt.figure(figsize = (12,8))
plt.suptitle('Subplots of Different Functions')     # To give entire figure a title, use suptitle() command.

plt.subplot(1,3,1)                                  # One row, three columns; select 1st subplot
plt.plot(nums, nums)
plt.title('y = x')

plt.subplot(1,3,2)                                  # One row, three columns; select 2nd subplot
plt.plot(nums, nums**2, c = 'r')
plt.title('y = x squared')

plt.subplot(1,3,3)                                  # One row, three columns; select 3rd subplot
plt.plot(nums, nums**3, c = 'y')
plt.title('y = x cubed')

plt.tight_layout()                                  # Ask matplotlib to take care of formatting for us.
plt.show()
```

```python
plt.figure(figsize = (8,8))
plt.suptitle('Different Functions on 2x2 figure', fontsize = 30)

plt.subplot(2,2,1)              # Two rows, two columns; first subplot
plt.plot(nums, nums)
plt.title('y = x')

plt.subplot(2,2,2)              # Two rows, two columns; second subplot
plt.plot(nums, nums**2)
plt.title('y = x^2')

plt.subplot(2,2,3)              # Two rows, two columns; third subplot
plt.plot(nums, nums**3)
plt.title('y = x^3')

plt.subplot(2,2,4)              # Two rows, two columns; fourth subplot
plt.plot(nums, nums**4)
plt.title('y = x^4')

plt.tight_layout()
plt.show()
```

```python
###WITHOUT SHARING Y AXIS ###


first_class = titanic[titanic['Pclass']==1]['Age']
second_class = titanic[titanic['Pclass']==2]['Age']
third_class = titanic[titanic['Pclass']==3]['Age']

plt.figure(figsize = (10,4))

plt.subplot(1,3,1)
plt.hist(first_class, label="1st Class", color="#f1c40f", bins=20)
plt.title('First Class')

plt.subplot(1,3,2)
plt.hist(second_class, label="2nd Class", color="#3498db", bins=20)
plt.title('Second Class')

plt.subplot(1,3,3)
plt.hist(third_class, label="3rd Class", color="#e74c3c", bins=20)
plt.title('Third Class')


plt.show()
```

```python
###SHARING Y AXIS


first_class = titanic[titanic['Pclass']==1]['Age']
second_class = titanic[titanic['Pclass']==2]['Age']
third_class = titanic[titanic['Pclass']==3]['Age']

plt.figure(figsize = (10,4))

ax = plt.subplot(1,3,1)                                                 # Assign axes of first subplot as variable ax
plt.hist(first_class, label="1st Class", color="#f1c40f", bins=20)
plt.title('First Class')

plt.subplot(1,3,2, sharey = ax)                                         # When creating second subplot, share y axis with ax.
plt.hist(second_class, label="2nd Class", color="#3498db", bins=20)
plt.title('Second Class')

plt.subplot(1,3,3, sharey = ax)                                         # When creating third subplot, share y axis with ax.
plt.hist(third_class, label="3rd Class", color="#e74c3c", bins=20)
plt.title('Third Class')


plt.show()
```
